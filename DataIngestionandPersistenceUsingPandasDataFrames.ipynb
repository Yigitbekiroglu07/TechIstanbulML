```python
# SQL Server
import sqlalchemy as sa #SQl'e bağlanmanın yolu, sorgulatma işi yine pandas ile yapılacak.
```

```python
#SQL'in kurulu olduğu sunucunun adı
SERVER = "DESKTOP-3FP500S"
DATABASE = "AdventureWorksLT2022"
USERNAME = "admin"
PASS = "123456"
DRIVER = "ODBC Driver 17 for SQL Server"
#Kaynak Türü --> MsSQL (Buradan veri okunacak)
#pyodbc --> Hangi Python modülüyle veri alacağımızı söyleriz.
#&trusted_connection=yes --> Kullanıcı adı ve şifresini bilmeden bağlantı kurmamızı sağlar.
conn_str = f"mssql+pyodbc://{SERVER}/{DATABASE}?driver={DRIVER}&trusted_connection=yes"
```

```python
# Bu obje ile bağlantıyı oluşturabiliriz.
conn_str
```

```python
#Bağlantı oluşturunca SQL Server'da sorgu çalıştırınca motor oluşuyor.
#SQL motoru oluşturmak için engine create etmemiz gerekir.
try:
    engine = sa.create_engine(conn_str)
    engine.connect()
    print("Connection kuruldu.")
except Exception as e:
    print("Hata bulundu, hatanın sebebi de şu: ",e)   
```

```python
#SQL'den veri okuma
query = "SELECT * FROM [AdventureWorksLT2022].[SalesLT].[Customer]"
```

```python
# İçerisine 2 paraemtre gönderilir: SQL'in çalışacağı komut ve connection bağlantısı gereklidir.
df_sql = pd.read_sql(query,engine)
```

```python
#SQL'den veri yazma
ogrenciler = pd.DataFrame({"Ad":["Ahsen","Aybüke","Ayşegül,""Aziz"],
                          "Notlar":[90,100,95,100]
                          })
ogrenciler.to_sql(name="Ogrenciler",con=engine,index=False, if_exists = "replace")
```

```python
yeni_ogrenciler = pd.DataFrame({"Ad":["Furkan","Ecodation","Kübra,""Salih"],
                          "Notlar":[90,100,95,100]
                          })
yeni_ogrenciler.to_sql(name="Students",con=engine,index=False, if_exists = "append")
```
