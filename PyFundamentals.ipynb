```python
if ders_notu > 90:
    print("Tebrikler, AA ile geçtiniz.")
elif ders_notu > 80:
    print("Tebrikler BB ile geçtiniz.")
elif ders_notu > 70:
    print("Tebrikler CC ile geçtiniz.")
elif ders_notu > 60:
    print("Tebrikler DD ile geçtiniz.")
else:
    print("Maalesef kaldınız")
```
