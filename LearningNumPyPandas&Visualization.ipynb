```python
#Merge --> Elimizde iki veya daha fazla df varsa bu iki veri kaynağını alt alta birleştirebiliriz.
#Bunu yapmak için SQL'deki Join mantığını bilmemiz gerekir.
```

```python
# Merge - Join

import pandas as pd

df_calisanlar = pd.DataFrame({
    "Calisan_ID": [101, 102, 103, 104],
    "Ad": ["Ali", "Veli", "Ayşe", "Fatma"],
    "BolumID": ["1", "1", "5", "3"]
})

df_bolumler = pd.DataFrame({
    "BolumID": ["1", "2", "3", "4"],
    "Bolum": ["IT", "HR", "Finance", "Legal"]
})
```

```python
merge_inner = pd.merge(df_calisanlar, df_bolumler, on="BolumID",how="inner")
```

```python
merge_inner # inner --> iki tablodada olma zorunluluğu var demektir, BolumID iki tarafta da olan 1 ve 3 değerleridir.
```

```python
merge_left = pd.merge(df_calisanlar, df_bolumler, on="BolumID",how="left")
```

```python
merge_left # çalışanlar tablosu olduğu gibi geldi. Sağdaki tablodan da bulabildiği kadar getirdi.
```

```python
merge_right = pd.merge(df_calisanlar, df_bolumler, on="BolumID",how="right")
```

```python
merge_right
```

```python
# Concat --> SQL'de Union olarak bilinir. Verileri alt alta birleştirir.
```

```python
df1 = pd.DataFrame({"Sira": [1, 2, 3]})
df2 = pd.DataFrame({"Sira": [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) #Dikeyde birleştirir.
```

```python
pd.concat([df1,df2], axis=1) #Yatayda birleştirir.
```

```python
#Dikeyde birleştirem fakat indexleri sıfırlama
pd.concat([df1,df2], axis=0, ignore_index=True)
```
