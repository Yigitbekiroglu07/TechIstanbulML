```python
class GidenAraba:
    def __init__(self,marka, model, yıl,motor,):
        self.marka = marka
        self.model = model 
        self.yıl = yıl
        self.motor = "kapalı"
        self.fren = "Aktif" # Değerini verdiğimiz için yukarıya parametre yollamak zorunda değiliz.
    def elfreni(self):
        self.fren = "Pasif"
    def calistir(self):
        print("Araba çalıştı, gür gür gür.")
        self.motor = "Açık"
        # İç içe method
        self.elfreni() # Motoru çalıştırdığında otomatik olarak el frenini indirmesini istedik.
    def kapat(self): 
        print("Araba kapatıldı")
        self.motor = "Kapalı"   
```

```python
#Encapsulation hali
class GüvenlikliBankaHesabı:
    def __init__(self,iban,isim):
        self.iban = iban
        self.__tutar = 0 #Encapsulation yaptık çünkü tutar değerinin dışardan değiştirilmesini istemiyoruz.
        self.isim = isim
    def havaleyap(self, miktar): #setter metodu (İçerideki saklı değeri (self.__tutar) dışardan göstermeye yarayan metod)
        self.__tutar = self.__tutar + miktar
        print("Para aktarımı tamamlandı, hesabınızdaki tutar: ", self.__tutar)
```

```python
güvenliklibankahesabı3 = GüvenlikliBankaHesabı("TR2000I457854754754","Alperen")
```

```python
güvenliklibankahesabı3.havaleyap(100)
```

Abstraction --> Kendisinden nesne türetilemez ama alt sınıflarından türetilebilir. Örneğin Araç diye bir class oluşturursak Araba ve Otobüs classları olduğunda Araç bir abstract sınıf olur.

```python
from abc import ABC, abstractmethod  # abc kütüphanesinden ABC classını import etmeliyiz ve yeni bir soyut class türetmek için bundan türetmeliyiz.
# classın soyut olması için ABC kullanılmalı.

class Araç(ABC):  # Araç classının soyut olması için ABC classından türetmeliyiz.
    @abstractmethod  # Araçtan türeyecek Araba ve Otobüste gitmek fonksiyonu zorunlu olması için bu işaretlemeyi yaparız.
    def gitmek(self):
        print("Araba gidiyor")

class Araba(Araç):
    def gitmek(self):
        print("Araba gidiyor")
```
